```python
#Splitting data into training set and testing set
# importing train test split from sklearn
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(carprice_dataframe_final11, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
#Importing MinMax scaler for scalling purpose
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the neumeric variable columns except the 'yes-no'(Binary) and 'dummy'(Categorical coulmn converted) variables
num_vars = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

```python
#Dividing into X and y sets for model building
y_train = df_train.pop('price')
X_train = df_train
```

```python
# Importing RFE and LinearRegression for model building
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
#columns would be splitted into true and false
#True columns are selected by RFE
#Numbers represent the level of preference
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#Array of columns which are supported by RFE for model building
col = X_train.columns[rfe.support_]
col
```

```python
#Array of columns which are not supported by RFE and having different levels of preference
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```
